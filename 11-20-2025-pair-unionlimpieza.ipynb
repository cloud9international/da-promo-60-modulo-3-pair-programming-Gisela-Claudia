2. Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:
   * No tengan espacios.
   * Estén en minúscula.
   * No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
   * Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
   * Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
df_wd_merge.head(2)
```

```python
df_wd_merge.columns
```

```python
import re

nuevas_columnas = {}

for columna in df_wd_merge.columns:
    # Pasar a minúsculas
    col_lower = columna.lower()
    # Reemplazar espacios y barras por guiones bajos
    col_clean = col_lower.replace(" ", "_").replace("/", "_").replace("-", "_")
    # Quitar paréntesis y su contenido con regex
    col_clean = re.sub(r"\s*\(.*?\)", "", col_clean)
    # Quitar dobles guiones bajos consecutivos si quedan
    col_clean = col_clean.replace("__", "_")
    # Quitar guiones bajos al principio o final (opcional)
    col_clean = col_clean.strip("_")

    nuevas_columnas[columna] = col_clean


nuevas_columnas
```

```python
df_wd_merge = df_wd_merge.rename(columns=nuevas_columnas)
```
