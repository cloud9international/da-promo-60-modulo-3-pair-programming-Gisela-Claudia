1. ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

```python
# (df.groupby("continent")["country"].count() / df["country"].sum()) * 100  - intento no funciona
```

```python
# Conteo de países por continente
continent_counts = df.groupby("continent")["country"].count()

# Total de países
total_countries = df["country"].count()

# Porcentaje por continente
percentage_by_continent = (continent_counts / total_countries) * 100

# Continente con más países
top_continent = percentage_by_continent.idxmax()
top_percentage = percentage_by_continent.max()

print(f"El continente con más países es {top_continent} con un {top_percentage:.2f}% del total.")
```

¿2. Cuáles son los 10 países con la mayor área forestal?

```python
df_top10 = df.sort_values(by="forested_area", ascending=False).head(10)     # IA - nos faltaba el by= antes de "forested_area"
print(df_top10[["country", "forested_area"]])
```

```python
sns.countplot(data=df_top10,x="forested_area")
plt.xticks(rotation=45)
plt.show() 
```

```python
# IA - eligio barplot y los parametros de abajo:
plt.figure(figsize=(12,6))
sns.barplot(data=df_top10, x="forested_area", y="country")
plt.title("Top 10 países con mayor área forestal")
plt.xlabel("Área forestal")
plt.ylabel("País")
plt.tight_layout()
plt.show()
```
